## Ideal simulation of a circuit with AerSimulator

```python
from qiskit import QuantumCircuit   # to define the quantum circuit
from qiskit_aer import AerSimulator # ideal simulator backend
from qiskit.visualization import plot_histogram  # to plot histograms of simulation results

```

### Circuit definition

```python
circ = QuantumCircuit(2, 2)
circ.x(0)
#circ.h(0)
#circ.cx(0, 1)
circ.measure([0, 1], [0, 1])
circ.draw(output='mpl')
```

### Use ideal simulator

```python
idealSim = AerSimulator()                   # create the ideal simulator
result_ideal = idealSim.run(circ).result()  # run the circuit in the simulator
counts_ideal = result_ideal.get_counts()    # get the counts

plot_histogram(result_ideal.get_counts())
```
